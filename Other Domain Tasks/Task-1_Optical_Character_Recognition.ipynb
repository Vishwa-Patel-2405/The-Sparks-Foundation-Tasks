## Import all required libraries

```python
import cv2
import matplotlib.pyplot as plt
import easyocr
from pylab import rcParams
import warnings
warnings.filterwarnings("ignore")
from IPython.display import Image
rcParams['figure.figsize'] = 10,20
```

## Downloading English language model

```python
Lang_reader = easyocr.Reader(['en'])
```

## Demo image preview

```python
Image('Img/Demo.jpeg')
```

## Detection of texts in Demo.jpeg

```python
output1 = Lang_reader.readtext('Img/Demo.jpeg')
```

## Print the detected information from texts in image

### This list shows bounding box co-ordinates, Text and Confidence level

```python
output1
```

## Extract the fetched output text & bounding box coordinates

```python
cord1 = output1[-2][0]
print(output1[-2][1])
```

## Minimum value of coordinates

```python
x_min1,y_min1 = [min(idx) for idx in zip(*cord1)]
```

## Maximum value of coordinates

```python
x_max1,y_max1 = [max(idx) for idx in zip(*cord1)]
```

## Visualization of bounding box co-ordinates of particular text

```python
image = cv2.imread('Img/Demo.jpeg')
cv2.rectangle(image,(int(x_min1),int(y_min1)),(int(x_max1),int(y_max1)),(0,255,0),2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
cord2 = output1[-1][0]
print(output1[-1][1])
```

```python
x_min2,y_min2 = [min(idx) for idx in zip(*cord2)]
x_max2,y_max2 = [max(idx) for idx in zip(*cord2)]
```
